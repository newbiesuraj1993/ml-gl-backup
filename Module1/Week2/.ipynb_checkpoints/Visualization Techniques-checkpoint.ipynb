```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
df=pd.read_csv('./automobile-dataset/Automobile.csv')
```

```python
df.head()
```

```python
sns.distplot(df['normalized_losses'])
plt.show()
```

```python
sns.distplot(df['normalized_losses'],kde=False,rug=True)
plt.show()
```

```python
sns.jointplot(df['engine_size'],df['horsepower'])
plt.show()
```

```python
sns.jointplot(df['engine_size'],df['wheel_base'],kind="hex") #hex can be replaced by residual or reg or kde(kernal density estimation)
plt.show()
```

```python
sns.pairplot(df[['engine_size','wheel_base','horsepower']])
```

```python
sns.stripplot(df['fuel_type'],df['horsepower'],jitter=True)
```

```python
sns.swarmplot(df['fuel_type'],df['horsepower'])
```

```python
sns.boxplot(df['fuel_type'],df['engine_size'],hue=df['number_of_doors'])
```

```python
sns.barplot(df['fuel_type'],df['engine_size'],hue=df['number_of_doors'])
```

```python
sns.countplot(df['body_style'],hue=df['engine_location'])
```

```python
sns.pointplot(df['fuel_system'],df['horsepower'],hue=df['number_of_doors'])
```

```python
sns.factorplot(x='fuel_type',y='horsepower',col='engine_location',hue='number_of_doors',data=df,kind='swarm')
```

```python
sns.factorplot(x='fuel_type',y='horsepower',col='make',hue='number_of_doors',data=df,kind='bar')
#point,bar,count,box,violin,strip
```
